```python
import pandas as pd
```

```python
df = pd.read_csv('riscv64instrdesc.csv',sep='=', header=None)
```

```python
df
```

```python
ext = df[0].str.extract('(?P<date>[a-zA-Z_]+)', expand=True)
```

```python
df.insert(0, "Extracted", ext, True)
```

```python
df['Extracted']
```

```python
df.to_csv('extracted_riscv64.csv',header=False)
```

```python
grouped = df.groupby(['Extracted'])
```
